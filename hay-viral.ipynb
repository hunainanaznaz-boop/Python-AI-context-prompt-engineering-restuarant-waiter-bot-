# PYTHON + AI CONTEXT PROMPT ENGINEERING (RESTUARANT WAITER BOT)

#  Summary

**AI Restaurant Waiter Bot is a Python-based chatbot with a Gradio interface that simulates a restaurant waiter. It greets customers, displays the menu, recommends food items, takes orders, validates menu items, maintains order history, calculates bills, and generates receipts while following strict menu constraints.**

## Key Features

* Interactive Restaurant Chatbot,
* Customer Greeting System,
* Menu Display,
* Food Recommendation,
* Menu Item Validation,
* Case-Insensitive Input Handling,
* Order Management, 
* Order History Tracking,
* Automatic Bill Calculation,
* Receipt Generation,
* Gradio-Based User Interface,
* Strict Menu Constraints.**


## Technologies Used
* Python
* Gradio
* Regex
* Dictionaries
* Prompt Engineering

## Code

```python
# ==========================================
# AI RESTAURANT WAITER BOT
# Python + Gradio + Prompt Engineering
# ==========================================

import gradio as gr
import re

# ==========================================
# SYSTEM PROMPT
# ==========================================

SYSTEM_PROMPT = """
You are a strict conversational automation waiter proxy agent.

Rules:
1. Only use items available in the active menu.
2. Do not suggest external items.
3. Validate every order against menu dictionary.
4. Maintain customer order context.
5. Generate receipt only from validated orders.

Strict Limit:
Do not hallucinate external entries.
"""

# ==========================================
# MENU CONTEXT
# ==========================================

menu = {
    "Chicken Pizza": 1200,
    "Beef Burger": 650,
    "Club Sandwich": 450,
    "Fries": 250,
    "Coke": 100,
    "Chicken Biryani": 400,
    "Mac n Cheese": 699,
    "Veggie Burger": 250,
    "Corn Dog": 150,
    "Shashlik With Rice": 2000,
    "Mango Shake": 200,
    "BBQ": 1500
}

# Case-insensitive lookup

menu_lookup = {
    item.lower(): item
    for item in menu
}

# ==========================================
# USER DATA
# ==========================================

customer_name = ""
order_history = {}
total_bill = 0

# ==========================================
# VALIDATE ITEM
# ==========================================

def validate_item(user_input):

    cleaned = re.sub(
        r"[^a-zA-Z\s]",
        "",
        user_input
    ).strip().lower()

    if cleaned in menu_lookup:
        return menu_lookup[cleaned]

    return None

# ==========================================
# SHOW MENU
# ==========================================

def get_menu():

    menu_text = "📋 OUR MENU\n\n"

    for item, price in menu.items():
        menu_text += f"🍽️ {item} - Rs.{price}\n"

    return menu_text

# ==========================================
# RECEIPT
# ==========================================

def generate_receipt():

    receipt = (
        "🧾 RECEIPT\n"
        "==============================\n"
        f"Customer: {customer_name}\n"
        "==============================\n\n"
    )

    for item, qty in order_history.items():

        subtotal = menu[item] * qty

        receipt += (
            f"{item}\n"
            f"{qty} x Rs.{menu[item]} = Rs.{subtotal}\n\n"
        )

    receipt += (
        "------------------------------\n"
        f"TOTAL BILL: Rs.{total_bill}\n"
        "==============================\n"
        "Thank you for visiting! 😊"
    )

    return receipt

# ==========================================
# CHATBOT LOGIC
# ==========================================

def restaurant_bot(message, history):

    global customer_name
    global order_history
    global total_bill

    user_message = message.lower().strip()

    # Take customer name

    if customer_name == "":

        customer_name = message.strip()

        return (
            f"Nice to meet you {customer_name}! 😊\n\n"
            f"{get_menu()}\n"
            "⭐ Today's Recommendation: Chicken Pizza\n"
            "Would you like to order something?"
        )

    # Show menu

    if "menu" in user_message:

        return get_menu()

    # Recommendation

    if "recommend" in user_message or "suggest" in user_message:

        return (
            "⭐ I recommend Chicken Pizza.\n"
            "It is our best selling item."
        )

    # Customer wants to continue

    if user_message == "yes":

        return (
            "😊 Sure!\n"
            "Please tell me which item you would like to order."
        )

    # Customer finished ordering

    if user_message in ["no", "done", "finish", "checkout"]:

        if not order_history:

            return "You have not ordered anything yet."

        return (
            "✅ Your order has been finalized.\n\n"
            "Type 'receipt' to view your bill."
        )

    # Receipt

    if "receipt" in user_message or "bill" in user_message:

        if not order_history:
            return "No items ordered yet."

        return generate_receipt()

    # Order Processing

    item = validate_item(message)

    if item:

        if item in order_history:
            order_history[item] += 1
        else:
            order_history[item] = 1

        total_bill += menu[item]

        return (
            f"✅ {item} added to your order.\n"
            f"Current Bill: Rs.{total_bill}\n\n"
            "Would you like to add anything else? (yes/no)"
        )

    # Default Response

    return (
        "❌ Sorry, that item is not available.\n\n"
        "Type 'menu' to see available items."
    )

# ==========================================
# WELCOME MESSAGE
# ==========================================

welcome_message = [
    (
        None,
        "👋 Welcome to Hunaina's Kitchen! 🍽️\n\n"
        "I am your virtual waiter.\n"
        "Please tell me your name to start 😊"
    )
]

# ==========================================
# GRADIO UI
# ==========================================

with gr.Blocks() as demo:

    gr.Markdown("# 🍽️ Hunaina's Kitchen")

    chatbot = gr.Chatbot(
        value=welcome_message,
        height=500
    )

    textbox = gr.Textbox(
        placeholder="Type your message...",
        label="Customer"
    )

    def chat(message, history):

        response = restaurant_bot(
            message,
            history
        )

        history.append(
            (message, response)
        )

        return "", history

    textbox.submit(
        chat,
        inputs=[textbox, chatbot],
        outputs=[textbox, chatbot]
    )

# ==========================================
# RUN APP
# ==========================================

demo.launch()
```
